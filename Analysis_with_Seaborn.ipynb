### Short word about the Star Tables
>The apparent magnitude is how bright a star is as viewed from Earth

>The absolute magnitude is how bright it *would* be if it was 10 parsecs (32.6 light years) from Earth

>Luminosity is *independent* of the distance from Earth

# <p style="text-align: center;">Joint Plot</p>


```python
##Try kinds = “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” 
sns.jointplot(x = 'Speed', y = 'Acceleration', hue = 'Toughness', data = simpsons, height = 10)

''' Toughness is negatively correlated with acceleration, and for obvious reasons because the 
stronger cars are a lot slower, and the missions are designed around this feature. Similarly, cars
that break a lot easier are also the fastest, which also factors into missions. 
The sweet spot seems to be around the speed/accel. = 4.5 and toughness = 3.5 stats of the Hearse.'''
```

# <p style="text-align: center;">Matrix Plots: Heat Maps</p>


```python
#First we need to square away the data's unnecessary indicies and get the data formatted into a matrix
#And if we use the correlation function we just find out how influential a variable is on our results
simpsons2 = simpsons.corr()
simpsons2

#Note about tese type of plots: The main diagonal is an axis of symmetry
#So the uppoer triangular matrix is a reflection of the lower triangular matrix
```

```python
sns.heatmap(simpsons2, cmap = 'Blues', linecolor = 'white', linewidth = 1, annot = True)
#Again, pretty straightforward map. Cars with the most speed have the most acceleration 
#and also the least toughness.
```

```python
ygo3 = ygo2.corr()
sns.heatmap(ygo3, annot = True, cmap='binary')
#Another heat map but this one is gray-scaled. This jsut shows a 3x3 correlation grid between
#various fields, such as correlation of monster level ("Level") vs Attack = 91%, 
#Which makes sense, higher level monsters will have higher attack
#Notice also that it is symmetrical about the main-diagonal
#which means this could realistically be represented by an upper- or lower-triangular matrix
```
